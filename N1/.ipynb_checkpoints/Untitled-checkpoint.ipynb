```python
final = ""
with open('12N1.txt', encoding="utf-8") as f:
    line = f.readlines()
for index in range(len(line)):
    string = ""
    cell = (line[index].replace("\n", "")).split(",")
    cell[-1] = cell[num3[index]+1]  ############################
    for cells in cell:
        string += cells
        string += ","
    final += string[:-1] + "\n"
# print(final)
```
